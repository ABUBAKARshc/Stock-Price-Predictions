```python
#Visualize this stock price data to get a clear picture of the increase and decrease stock price of Suzlon:
figure = go.Figure(data = [go.Candlestick(x = data["Date"],
                                          open = data["Open"], high = data["High"],
                                         low = data["Low"], close = data["Close"])])

figure.update_layout(title = "Suzlone Stock price Analysis", xaxis_rangeslider_visible = False)
figure.show()
```

### Looking at the correlation between the features in this dataset

```python
print(data.drop(["Date"],axis=1).corr())
```

```python
from autots import AutoTS
```

```python
model = AutoTS(forecast_length = 5, frequency = "infer", ensemble = "simple")
model = model.fit(data, date_col = "Date", value_col = "Close", id_col = None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```
